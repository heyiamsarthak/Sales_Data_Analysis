```python
#import libraries
import pandas as pd
import os
```

```python
##save all months data in files
files = [file for file in os.listdir(r'C:\Users\sarth\Downloads\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data')]

#create an empty dataframe
all_months_data = pd.DataFrame()

#concatenate all 12 months data into single df
for file in files:
    df = pd.read_csv(r'C:\Users\sarth\Downloads\Pandas-Data-Science-Tasks-master\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\\'+file)
    all_months_data = pd.concat([all_months_data,df])

#print 
all_months_data    
```

# Cleaning The Data To remove NaN Values

```python
#Drop NaN values
all_months_data = all_months_data.dropna()
```

```python
all_months_data
```

```python
#Extract Month from Order Date column
all_months_data['Month'] = all_months_data['Order Date'].str[:2]
```

```python
#Change Month to numeric while converting the non numeric to Nan
all_months_data['Month'] = pd.to_numeric(all_months_data['Month'], errors='coerce')
```

```python
#Removing the NaN rows arised from last cell
all_months_data = all_months_data.dropna()
```

```python
#Converting Month into int
all_months_data['Month'] = all_months_data['Month'].astype('int32')
all_months_data
```

```python
#Converting two cols from string to int and float
all_months_data['Quantity Ordered'] = all_months_data['Quantity Ordered'].astype('int32')
all_months_data['Price Each'] = all_months_data['Price Each'].astype('float')

#making a new col 'Revenue' to record the Sales
all_months_data['Revenue'] = all_months_data['Price Each']*all_months_data['Quantity Ordered']
all_months_data
```

# What was the best month for sales?

```python
#Group by Months
temp1 = all_months_data.groupby('Month').sum()
```

```python
temp1
```

```python
#Visualization
import matplotlib.pyplot as plt

mon = range(1,13)
plt.bar(mon , temp1['Revenue'])

plt.xlabel('Months')
plt.xticks(mon)
plt.ylabel('Sales')
plt.title('Sales per Month')
```

# Which city sold the most products?

```python
#splitting address into street city and state
all_months_data[['Street', 'City' , 'State']] = all_months_data['Purchase Address'].str.split(',' , expand = True)

all_months_data
```

```python
#Split state into state and zip using regex 
all_months_data[['State', 'ZIP']] = all_months_data['State'].str.extract(r'\s([A-Z]{2})\s(\d{5})')
```

```python
#grouped by city AND state because there can be two cities with the same name. Here, it is Portland
group_cols = ['City','State']
temp2 = all_months_data.groupby(group_cols).sum()
```

```python
#Keep only the revenue column(city and state cols are indexes)
temp2 = temp2['Revenue']
```

```python
temp2
```

```python
temp2 = temp2.sort_values(ascending = False)
temp2
```

```python
#resetting the index so that city and state become normal cols
temp2 = temp2.reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(temp2['City'] + ', ' + temp2['State'], temp2['Revenue'], color='skyblue')
plt.xlabel('City, State')
plt.ylabel('Revenue')
plt.title('Revenue by City and State')
plt.xticks(rotation=45)#Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# What products are most often sold together?

```python
df =all_months_data [all_months_data['Order ID'].duplicated(keep = False)]
df
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df
```

```python
df = df[['Order ID' , 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```
